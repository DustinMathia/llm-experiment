```python
# function that prints colored text over
def print_color(text, k=0):
    # ANSI color codes for various colors
    global colors 
    colors= [
        '\033[90m', # Bright Black
        '\033[91m', # Bright Red
        '\033[92m', # Bright Green
        '\033[93m', # Bright Yellow
        '\033[94m', # Bright Blue
        '\033[95m', # Bright Magenta
        '\033[96m', # Bright Cyan
        '\033[97m', # Bright White
        
    ]
    reset_code = '\033[0m' # Resets the color after the text
    k_color = colors[k % len(colors)] # iterate colors
    print(f"{k_color}{text}{reset_code}")
```

```python
# example of how colors repeat
for i in range(20):
    print_color(f"{i}: This text is color #{i % 8}", i) #there are 8 colors in private list
```

```python
import asyncio

async def prompter(k):
    print_color(f"{k} Entry", k)
    await asyncio.sleep(1)
    print_color(f"{k} Exit", k)

await prompter(2)
```

Async code

```python
import asyncio

async def main():
    L = await asyncio.gather(*[prompter(i) for i in range(20)], return_exceptions=True)
    #print(L)
    

await main()
```

```python
import asyncio

async def main():
    tasks = [prompter(i) for i in range(20)]

    for future in asyncio.as_completed(tasks):
        result = await future
    

await main()
```

```python
import asyncio

sem = asyncio.Semaphore(5)

async def prompter(k):
    async with sem:
        print_color(f"{k} Entry", k)
        await asyncio.sleep(1)
        print_color(f"{k} Exit", k)

async def main():
    tasks = [prompter(i) for i in range(20)]

    for future in asyncio.as_completed(tasks):
        result = await future
    

await main()
```
